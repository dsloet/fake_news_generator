## Create news sources

This part creates both fake and real sources which can then be used by the bots to post 

```python
import newspaper
```

```python
cnn_paper = newspaper.build('http://cnn.com', memoize_articles=True)
titles = []
for articles in cnn_paper.articles:
    try:
        articles.download()
        articles.parse()
        #print(articles.title)
        titles.append(articles.title)
    except:
        pass
```

```python
titles
```

```python
copy_titles = titles.copy()
```

That was real news, now fake:


```python
import markovify
```

```python
with open('data2.txt', 'r') as f:
    text = f.read()
```

```python
text_model = markovify.Text(text)
```

## This part will generate fake news

The accounts.csv file contains a fake or not fake boolean. This will be used now to generate fake news articles using the Markovify technique

```python
df_fake = accounts.loc[accounts.fake == 1]
```

```python
df_fake
```

Now we loaded the data2.txt file into text_model and we have the fake news generating bots:

We can use that to create short sentences

```python
driver = webdriver.Chrome(executable_path = 'chromedriver.exe')
driver.get('http://127.0.0.1:5000/')
for index, row in df_fake.iterrows():
    # username in the form
    username = driver.find_element_by_id("username")
    
    # password in the form
    password = driver.find_element_by_id("password")
    
    # Get user from csv file
    user = row['username']
    
    # Get password from csv file
    pwd = row['password']
    
    # Send username and password to sign in screen
    username.send_keys(user)
    password.send_keys(pwd)
    
    driver.find_element_by_name("submit").click()
    time.sleep(2)
    
    # we are now logged in.
    
    # Create fake news
    post_location = driver.find_element_by_id("post")
    post = text_model.make_short_sentence(140)
    post_location.send_keys(post)
    time.sleep(1)
    driver.find_element_by_name("submit").click()
    time.sleep(1)
    driver.find_element_by_partial_link_text('Logout').click()
```
